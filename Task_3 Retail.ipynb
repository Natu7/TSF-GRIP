```python
data = data.drop(['Country', 'City', 'Postal Code'], axis=1)
```

```python
data.drop_duplicates(subset = None, inplace = True)
data.duplicated().sum()
```

```python
corelation = data.corr()
sns.heatmap(corelation, annot=True)
```

```python
sns.pairplot(data)
```

```python
sns.relplot(x = 'Sales', y = 'Profit', hue='Category', data=data)
```

```python
sns.histplot(x = 'Category', y = 'Sales', data=data)
```

```python
sns.catplot(x='Profit', kind='box',  data=data)
```

```python
sns.catplot(x='Sales', kind='box',  data=data)
```

```python
sns.boxplot(x='Category',y='Profit',data=data)
```

```python
sns.boxplot(x='Category',y='Profit',data=data)
```

```python
sns.relplot(x = 'Sales', y = 'Profit', hue='Discount', data=data)
```

```python
plt.figure(figsize = (12,4))
sns.set(font_scale=1, palette= "viridis")
sns.barplot(x = "Region",y = "Profit" ,hue = "Segment", data = data)

```

```python
Top10_Sales = data.groupby("State").Sales.sum().nlargest(n =10)
Top10_Profits = data.groupby("State").Profit.sum().nlargest(n =10)
```

```python
Top10_Sales.index
```

```python
Top10_Profits.index
```

```python
plt.style.use('seaborn')
Top10_Sales.plot(kind ='bar', figsize =(14,8), fontsize =14)
plt.xlabel("States", fontsize =13)
plt.ylabel("Total Sales",fontsize =13)
plt.title("Top 10 States by Sales",fontsize =16)
plt.show()
```

```python
plt.style.use('seaborn')
Top10_Profits.plot(kind ='bar', figsize =(14,8), fontsize =14)
plt.xlabel("States", fontsize =13)
plt.ylabel("Total Profits",fontsize =13)
plt.title("Top 10 States by Profits",fontsize =16)
plt.show()
```

```python
swpl = data['State'].value_counts().nlargest(10)
swpl #Here, swpl stands for State wise profit and loss
data.groupby('State')[['Profit','Sales']].sum().plot.bar(figsize = (10,10))
```
